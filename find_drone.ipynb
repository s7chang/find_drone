```python
import os
import shutil

# 1. ÏûÑÏãú ÎîîÎ†âÌÜ†Î¶¨Ïóê Git Ï†ÄÏû•ÏÜå ÌÅ¥Î°†
temp_dir = "./temp_git_clone"
repo_url = "https://github.com/s7chang/find_drone.git"

if os.path.exists(temp_dir):
    shutil.rmtree(temp_dir)  # Í∏∞Ï°¥ ÏûÑÏãú ÎîîÎ†âÌÜ†Î¶¨Í∞Ä ÏûàÏúºÎ©¥ ÏÇ≠Ï†ú

os.system(f"git clone {repo_url} {temp_dir}")

# 2. ÌïÑÏöîÌïú ÌååÏùºÏùÑ ÌòÑÏû¨ ÎîîÎ†âÌÜ†Î¶¨Î°ú Î≥µÏÇ¨
for item in os.listdir(temp_dir):
    source = os.path.join(temp_dir, item)
    destination = os.path.join(".", item)

    if os.path.isdir(source):
        shutil.copytree(source, destination, dirs_exist_ok=True)  # ÎîîÎ†âÌÜ†Î¶¨ Î≥µÏÇ¨
    else:
        shutil.copy2(source, destination)  # ÌååÏùº Î≥µÏÇ¨

print("Git Ï†ÄÏû•ÏÜåÏùò ÌååÏùºÏùÑ ÌòÑÏû¨ ÎîîÎ†âÌÜ†Î¶¨Î°ú Î≥µÏÇ¨ÌñàÏäµÎãàÎã§.")

# 3. ÏûÑÏãú ÎîîÎ†âÌÜ†Î¶¨ ÏÇ≠Ï†ú
shutil.rmtree(temp_dir)
```
