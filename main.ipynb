```python
# Step 1: Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Clone the Git repository
!git clone https://github.com/s7chang/find_drone.git

# Step 3: Extract the dataset from Google Drive
import os
import tarfile

# Define paths
google_drive_dataset_path	= '/content/drive/My Drive/drone_data_set.tar'  # Update with actual path
dataset_extract_path		= './drone_data_set/'

# Extract dataset
if os.path.exists(google_drive_dataset_path):
    with tarfile.open(google_drive_dataset_path, 'r') as tar:
        tar.extractall(dataset_extract_path)
    print("Dataset extracted successfully.")
else:
    print(f"Dataset file not found at {google_drive_dataset_path}.")

# Change directory to the cloned repository
os.chdir('./find_drone')

# Import DroneSetup class
from set_yolo import DroneSetup

# Initialize the DroneSetup class
dir_data_set				= '../drone_data_set/'  # Use extracted dataset path
drone_setup					= DroneSetup(dir_data_set=dir_data_set)

# # Download training videos
# training_videos	= drone_setup.download_training_videos(image_video_urls)
# print("Downloaded training videos:", training_videos)

# # Extract frames from training videos
# drone_setup.extract_frames(training_videos)
# print("Extracted frames from videos.")

# # Download test video
# test_video_path	= drone_setup.download_test_video(test_video_url)
# print("Downloaded test video at:", test_video_path)

# # Create YAML file
# drone_setup.create_yaml()
# print("YAML file created.")

# Fine-tune the YOLO model
drone_setup.fine_tune_model(epochs=100, batch_size=32, optimizer='Adam', patience=5, img_size=640)
print("YOLO model fine-tuned.")

# Detect and draw results on test video
drone_setup.detect_and_draw()
print("Detection and drawing complete.")
```

```python
# Move the best.pt file and output video to Google Drive
import os
import shutil

# Define paths
google_drive_best_model_path = '/content/drive/My Drive/best.pt'  # Update with actual path
google_drive_output_video_path = '/content/drive/My Drive/output_video.mp4'  # Update with actual path

# Move best.pt
if os.path.exists(drone_setup.best_model_path):
    shutil.move(drone_setup.best_model_path, google_drive_best_model_path)
    print(f"best.pt file moved to {google_drive_best_model_path}.")
else:
    print(f"best.pt file not found at {drone_setup.best_model_path}.")

# Move output_video.mp4
if os.path.exists(drone_setup.output_video_path):
    shutil.move(drone_setup.output_video_path, google_drive_output_video_path)
    print(f"output_video.mp4 file moved to {google_drive_output_video_path}.")
else:
    print(f"output_video.mp4 file not found at {drone_setup.output_video_path}.")
```
